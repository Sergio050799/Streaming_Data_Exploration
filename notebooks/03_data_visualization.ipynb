# Visualización de datos

## 0.1 Librerías 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1.1 Carga de datos limpios

```python
all_data = pd.read_csv('../data/processed/all_data_cleaned.csv')
```

## 2.1 Visualizaciones

```python
sns.set(style='whitegrid')
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=all_data, x='platform', order=all_data['platform'].value_counts().index)
plt.title('Distribución de Contenido por Plataforma')
plt.xlabel('Plataforma')
plt.ylabel('Cantidad de Títulos')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
top_genres = all_data['genres'].value_counts().nlargest(15)
sns.barplot(x=top_genres.values, y=top_genres.index)
plt.title('Top 15 Géneros Más Comunes')
plt.xlabel('Cantidad de Títulos')
plt.ylabel('Género')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=all_data, x='platform', y='imdbAverageRating')
plt.title('Distribución de Calificaciones (IMDb) por Plataforma')
plt.xlabel('Plataforma')
plt.ylabel('Calificación Promedio (IMDb)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(all_data['releaseYear'], bins=30, kde=True)
plt.title('Distribución de Años de Lanzamiento')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Cantidad de Títulos')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=all_data, x='platform', y='imdbNumVotes')
plt.ylim(0, 50000)  # Limitar el eje Y para evitar outliers extremos
plt.title('Distribución de Votos (IMDb) por Plataforma')
plt.xlabel('Plataforma')
plt.ylabel('Número de Votos (IMDb)')
plt.xticks(rotation=45)
plt.show()
```
