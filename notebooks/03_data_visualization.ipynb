## 2.0 Visualizaciones

### Distribución del contenido

```python
# Contar las plataformas
platform_counts = Toda_Data_Limpia['Plataforma'].value_counts()

# Colores por plataforma
platform_colors = {'Netflix': 'red', 'Apple TV': 'gray', 'Amazon Prime Video': 'blue', 'HBO Max': 'purple'}

# Gráfico de dona
plt.figure(figsize=(8, 8))
plt.pie(
    platform_counts, 
    labels=[f"{platform} ({count})" for platform, count in zip(platform_counts.index, platform_counts)], 
    autopct='%1.1f%%', 
    startangle=90, 
    colors=[platform_colors.get(plat, 'black') for plat in platform_counts.index],  # Añadí un color por defecto
    wedgeprops={'linewidth': 1, 'edgecolor': 'white'}
)
plt.gca().add_artist(plt.Circle((0, 0), 0.70, color='white', fc='white'))  # Agujero en el centro
plt.title('Distribución de títulos por plataforma', fontsize=16)
plt.show()
```

```python
# Asegurarnos de que todas las plataformas están en el diccionario de colores
platform_colors = {
    'Netflix': 'red',
    'Amazon Prime Video': 'blue',
    'HBO Max': 'purple',
    'Apple TV+': 'gray'
}


# Procesar los datos: extraer porcentaje de uso por continente
usage_data = (
    Suscriptores['Porcentaje de Uso por Continente']
    .str.split(', ', expand=True)
    .stack()
    .reset_index(level=1, drop=True)  # Corregir el MultiIndex
    .str.split(': ', expand=True)
)
usage_data.columns = ['Continente', 'Porcentaje']
usage_data['Porcentaje'] = usage_data['Porcentaje'].str.replace('%', '').astype(float)
usage_data['Plataforma'] = Suscriptores.loc[usage_data.index, 'Plataforma']

# Preparar los datos para el gráfico
usage_pivot = usage_data.pivot_table(index='Continente', columns='Plataforma', values='Porcentaje', aggfunc='mean')

# Ajustar colores según las plataformas presentes
platform_colors_filtered = {platform: platform_colors[platform] for platform in usage_pivot.columns if platform in platform_colors}

# Gráfico de barras agrupadas
usage_pivot.plot(kind='bar', figsize=(12, 8), color=[platform_colors_filtered[col] for col in usage_pivot.columns])
plt.title('Porcentaje de suscriptores por plataforma en los diferentes continentes', fontsize=16)
plt.xlabel('Continente', fontsize=14)
plt.ylabel('Porcentaje de suscriptores (%)', fontsize=14)
plt.legend(title='Plataforma', loc='upper right')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Numero de titulos por plataforma

# Contar los títulos por plataforma
platform_counts = Toda_Data_Limpia['Plataforma'].value_counts()

# Crear el DataFrame para el gráfico
df = pd.DataFrame({'Plataforma': platform_counts.index, 'Número de Títulos': platform_counts.values})

# Colores por plataforma
platform_colors = {'Netflix': 'red', 'Apple TV': 'gray', 'Amazon Prime Video': 'blue', 'HBO Max': 'purple'}
colors = [platform_colors[platform] for platform in df['Plataforma']]

# Crear el gráfico de barras apiladas
plt.figure(figsize=(10, 6))
df.plot(kind='bar', x='Plataforma', y='Número de Títulos', legend=False, color=colors, stacked=True)

plt.title('Número de Títulos por Plataforma', fontsize=16)
plt.xlabel('Plataforma', fontsize=14)
plt.ylabel('Número de Títulos', fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Proporcion de contenido

content_type_counts = Toda_Data_Limpia['Tipo'].value_counts()

# Crear el gráfico de pastel
plt.figure(figsize=(8, 8))
plt.pie(
    content_type_counts, 
    labels=content_type_counts.index, 
    autopct='%1.1f%%', 
    startangle=90, 
    colors=['Red', 'Blue'],  
    wedgeprops={'linewidth': 1, 'edgecolor': 'white'}
)
plt.title('Proporción de Contenido por Tipo', fontsize=16)
plt.show()
```

```python
# Distribución de generos

genre_platform_counts = Toda_Data_Limpia.groupby(['Géneros', 'Plataforma']).size().unstack(fill_value=0)

# Ordenar los géneros por el número total de títulos
top_genres = genre_platform_counts.sum(axis=1).sort_values(ascending=False).head(5)

# Filtrar el DataFrame para obtener solo los géneros más consumidos
top_genre_platform_counts = genre_platform_counts.loc[top_genres.index]

# Colores por plataforma
platform_colors = {'Netflix': 'red', 'Apple TV': 'gray', 'Amazon Prime Video': 'blue', 'HBO Max': 'purple'}

# Crear el gráfico de barras apiladas en horizontal
top_genre_platform_counts.plot(kind='barh', stacked=True, color=[platform_colors[platform] for platform in top_genre_platform_counts.columns], figsize=(10, 8))

plt.title('Top 5 Géneros Más Consumidos por Plataforma', fontsize=16)
plt.xlabel('Número de Títulos', fontsize=14)
plt.ylabel('Género', fontsize=14)
plt.legend(title='Plataforma', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.gca().invert_yaxis()  # Invertir el eje y para mostrar de mayor a menor
plt.show()
```

### Calidad del contenido

```python
# Selección de columnas numéricas para calcular la correlación
numeric_columns = Toda_Data_Limpia[['Año de lanzamiento', 'Calificación promedio de IMDb', 'Número de votos en IMDb']]

# Matriz de correlación
correlation_matrix = numeric_columns.corr()

# Gráfico de mapa de calor
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Mapa de correlación entre variables', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
#¿Cuál es la calificación promedio de IMDb para cada plataforma?
# Colores por plataforma
platform_colors = {'Netflix': 'red', 'Apple TV': 'gray', 'Amazon Prime Video': 'blue', 'HBO Max': 'purple'}

# Crear el DataFrame para el gráfico
df_ratings = Toda_Data_Limpia[['Plataforma', 'Calificación promedio de IMDb']]

# Crear el gráfico de caja y bigotes
plt.figure(figsize=(12, 8))
sns.boxplot(
    x='Plataforma', 
    y='Calificación promedio de IMDb', 
    data=df_ratings, 
    palette=platform_colors
)

plt.title('Calificación promedio de IMDb por Plataforma', fontsize=16)
plt.xlabel('Plataforma', fontsize=14)
plt.ylabel('Calificación promedio de IMDb', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# ¿Qué plataformas tienen más títulos con altas calificaciones?
# Filtrar los títulos con calificaciones altas (>= 8)
high_ratings = Toda_Data_Limpia[Toda_Data_Limpia['Calificación promedio de IMDb'] >= 8]

# Contar los títulos con altas calificaciones por plataforma
high_ratings_counts = high_ratings['Plataforma'].value_counts()

# Colores por plataforma
platform_colors = {'Netflix': 'red', 'Apple TV': 'gray', 'Amazon Prime Video': 'blue', 'HBO Max': 'purple'}
colors = [platform_colors[platform] for platform in high_ratings_counts.index]

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
high_ratings_counts.plot(kind='bar', color=colors)

plt.title('Número de Títulos con Altas Calificaciones por Plataforma', fontsize=16)
plt.xlabel('Plataforma', fontsize=14)
plt.ylabel('Número de Títulos con Calificaciones >= 8', fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#
# Colores por plataforma
platform_colors = {'Netflix': 'red', 'Apple TV': 'gray', 'Amazon Prime Video': 'blue', 'HBO Max': 'purple'}

# Crear el DataFrame para el gráfico
df_votes_ratings = Toda_Data_Limpia[['Plataforma', 'Calificación promedio de IMDb', 'Número de votos en IMDb']]

# Crear el gráfico de dispersión con regresión opcional para cada plataforma
plt.figure(figsize=(12, 8))

# Graficar para cada plataforma
for platform, color in platform_colors.items():
    subset = df_votes_ratings[df_votes_ratings['Plataforma'] == platform]
    sns.regplot(
        x='Número de votos en IMDb', 
        y='Calificación promedio de IMDb', 
        data=subset, 
        scatter_kws={'s': 20}, 
        line_kws={'color': color}, 
        color=color, 
        label=platform
    )

plt.title('Relación entre la Cantidad de Votos y las Calificaciones por Plataforma', fontsize=16)
plt.xlabel('Número de votos en IMDb', fontsize=14)
plt.ylabel('Calificación promedio de IMDb', fontsize=14)
plt.legend(title='Plataforma')
plt.grid(True)
plt.show()
```

### Alcance geográfico

```python
# Procesar datos: calcular la cobertura global por plataforma
coverage = Toda_Data_Limpia.groupby('Plataforma')['Países disponibles'].apply(lambda x: x.str.split(',').explode().nunique())

# Gráfico de barras horizontal
plt.figure(figsize=(10, 6))
coverage.sort_values(ascending=True).plot(kind='barh', color=[platform_colors[plat] for plat in coverage.index])
plt.title('Cobertura global (número de países) por plataforma', fontsize=16)
plt.xlabel('Número de países', fontsize=14)
plt.ylabel('Plataforma', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Añadir el color de 'Apple TV+'
platform_colors = {
    'Netflix': 'red',
    'Amazon Prime Video': 'blue',
    'HBO Max': 'purple',
    'Apple TV+': 'gray'
}


# Procesar datos: extraer regiones y suscriptores
region_data = Suscriptores['Suscriptores por Región'].str.split(', ', expand=True).stack()
region_data = region_data.str.split(': ', expand=True).reset_index(drop=True)
region_data.columns = ['Región', 'Suscriptores']
region_data['Suscriptores'] = region_data['Suscriptores'].str.replace('M', '').astype(float)
region_data['Plataforma'] = Suscriptores.loc[region_data.index // 4, 'Plataforma'].values

# Pivotar datos para gráfico apilado
stacked_data = region_data.pivot_table(index='Región', columns='Plataforma', values='Suscriptores', aggfunc='sum')

# Gráfico de barras apiladas
stacked_data.plot(kind='bar', stacked=True, figsize=(12, 8), color=[platform_colors[col] for col in stacked_data.columns])
plt.title('Distribución de suscriptores por región y plataforma', fontsize=16)
plt.xlabel('Región', fontsize=14)
plt.ylabel('Suscriptores (millones)', fontsize=14)
plt.legend(title='Plataforma', loc='upper right')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

### Precios y Caracteristicas

```python
#
# Gráfico de caja y bigotes
plt.figure(figsize=(10, 6))
sns.boxplot(x='Plataforma', y='Precio Actual', data=Precios, palette=platform_colors)
plt.title('Variación de precios actuales entre plataformas', fontsize=16)
plt.xlabel('Plataforma', fontsize=14)
plt.ylabel('Precio Actual ($)', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Agrupar datos por plataforma y características
feature_data = Precios.groupby(['Plataforma', 'Resolución Máxima', 'Reproducción simultánea'])['Precio Actual'].mean().reset_index()

# Gráfico de barras agrupado
plt.figure(figsize=(12, 8))
sns.barplot(x='Resolución Máxima', y='Precio Actual', hue='Plataforma', data=feature_data, palette=platform_colors)
plt.title('Relación entre características y precio por plataforma', fontsize=16)
plt.xlabel('Resolución Máxima', fontsize=14)
plt.ylabel('Precio Promedio ($)', fontsize=14)
plt.legend(title='Plataforma', loc='upper right')
plt.tight_layout()
plt.show()
```

```python
# Procesar datos: Contar la cantidad de planes con diferentes opciones
plan_data = Precios.groupby(['Plataforma', 'Tiene Plan Anual', 'Publicidad'])['Precio Actual'].count().unstack(fill_value=0)

# Gráfico de barras apiladas
plan_data.plot(kind='bar', stacked=True, figsize=(12, 8), color=['Red', 'blue']) 
plt.title('Impacto de planes anuales y publicidad en las plataformas', fontsize=16)
plt.xlabel('Plataforma', fontsize=14)
plt.ylabel('Cantidad de planes', fontsize=14)
plt.legend(title='Características', loc='upper right')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Competitivdad 

```python
# Calcular la cantidad de géneros únicos por plataforma
genre_diversity = Toda_Data_Limpia.groupby('Plataforma')['Géneros'].apply(lambda x: len(set(','.join(x).split(',')))).sort_values()

# Gráfico de barras horizontal
plt.figure(figsize=(10, 6))
genre_diversity.plot(kind='barh', color=[platform_colors[plat] for plat in genre_diversity.index])
plt.title('Diversidad de géneros por plataforma', fontsize=16)
plt.xlabel('Cantidad de géneros únicos', fontsize=14)
plt.ylabel('Plataforma', fontsize=14)
plt.tight_layout()
plt.show()
```
