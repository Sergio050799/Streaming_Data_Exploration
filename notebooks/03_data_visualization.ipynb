# Visualización de datos

## 0.1 Librerías 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1.1 Carga de datos limpios

```python
all_data = pd.read_csv('../data/external/all_data_with_continents.csv')
prices_data = pd.read_csv('../data/processed/streaming_platforms_prices_cleaned.csv')
```

## 2.1 Visualizaciones

```python
sns.set(style='whitegrid')
```

```python
# Visualización 1: Distribución de contenido por plataforma
plt.figure(figsize=(10, 6))
sns.countplot(data=all_data, x='platform', order=all_data['platform'].value_counts().index)
plt.title('Distribución de Contenido por Plataforma')
plt.xlabel('Plataforma')
plt.ylabel('Cantidad de Títulos')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualización 2: Distribución de géneros
plt.figure(figsize=(12, 8))
top_genres = all_data['genres'].value_counts().nlargest(15)
sns.barplot(x=top_genres.values, y=top_genres.index)
plt.title('Top 15 Géneros Más Comunes')
plt.xlabel('Cantidad de Títulos')
plt.ylabel('Género')
plt.show()
# Respuesta: ¿Qué géneros dominan en el catálogo general?
```

```python
# Visualización 3: Calificaciones promedio por plataforma
plt.figure(figsize=(10, 6))
sns.boxplot(data=all_data, x='platform', y='imdbAverageRating')
plt.title('Distribución de Calificaciones (IMDb) por Plataforma')
plt.xlabel('Plataforma')
plt.ylabel('Calificación Promedio (IMDb)')
plt.xticks(rotation=45)
plt.show()
# Respuesta: ¿Qué plataforma tiene contenido mejor calificado?
```

```python
# Visualización 4: Años de lanzamiento del contenido
plt.figure(figsize=(12, 6))
sns.histplot(all_data['releaseYear'], bins=30, kde=True)
plt.title('Distribución de Años de Lanzamiento')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Cantidad de Títulos')
plt.show()
# Respuesta: ¿En qué periodo se produjo más contenido?
```

```python
# Visualización 5: Popularidad (Número de Votos) por Plataforma
plt.figure(figsize=(10, 6))
sns.boxplot(data=all_data, x='platform', y='imdbNumVotes')
plt.ylim(0, 50000)  # Limitar el eje Y para evitar outliers extremos
plt.title('Distribución de Votos (IMDb) por Plataforma')
plt.xlabel('Plataforma')
plt.ylabel('Número de Votos (IMDb)')
plt.xticks(rotation=45)
plt.show()
# Respuesta: ¿Qué plataforma tiene títulos más populares?
```

```python
# Visualización 6: Precios actuales mensuales por plataforma y versión
plt.figure(figsize=(10, 6))
sns.barplot(data=prices_data, x='Plataforma', y='Precio Actual - Mensual', hue='Versión')
plt.title('Precios Actuales Mensuales por Plataforma y Versión')
plt.xlabel('Plataforma')
plt.ylabel('Precio Actual - Mensual ($USD)')
plt.xticks(rotation=45)
plt.legend(title='Versión')
plt.show()
# Respuesta: ¿Cómo varían los precios según el plan y la plataforma?
```

```python
# Preparar datos para el gráfico de línea
release_year_counts = all_data.groupby(['releaseYear', 'platform']).size().reset_index(name='count')

# Visualización 7: Evolución de contenido lanzado por año y plataforma
plt.figure(figsize=(12, 6))
sns.lineplot(data=release_year_counts, x='releaseYear', y='count', hue='platform')
plt.title('Evolución de Títulos Lanzados por Año y Plataforma')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Cantidad de Títulos')
plt.show()
```

```python
# Visualización 8: Relación entre votos y calificación
plt.figure(figsize=(10, 6))
sns.scatterplot(data=all_data, x='imdbNumVotes', y='imdbAverageRating', hue='platform', alpha=0.7)
plt.title('Relación entre Votos y Calificación por Plataforma')
plt.xlabel('Número de Votos (IMDb)')
plt.ylabel('Calificación Promedio (IMDb)')
plt.show()
# Respuesta: ¿Existe correlación entre popularidad (votos) y calificación?
```

```python
# Visualización 9: Heatmap de correlación entre variables
plt.figure(figsize=(10, 8))
correlation_matrix = all_data[['imdbAverageRating', 'imdbNumVotes', 'releaseYear']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Mapa de Correlación entre Variables')
plt.show()
# Respuesta: ¿Qué variables tienen correlaciones significativas?
```

```python
country_to_continent = {
    'US': 'North America', 'CA': 'North America', 'MX': 'North America', 
    'BR': 'South America', 'AR': 'South America', 'CO': 'South America',
    'DE': 'Europe', 'FR': 'Europe', 'GB': 'Europe', 'IT': 'Europe', 'ES': 'Europe',
    'CN': 'Asia', 'JP': 'Asia', 'IN': 'Asia', 'AU': 'Oceania', 'NZ': 'Oceania',
    'ZA': 'Africa', 'EG': 'Africa', 'NG': 'Africa', 'RU': 'Europe', 'TR': 'Asia',
}

countries_expanded = all_data.assign(
    availableCountries=all_data['availableCountries'].str.split(',')
).explode('availableCountries')

countries_expanded['continent'] = countries_expanded['availableCountries'].str.strip().map(country_to_continent)

# 2. Contar títulos por plataforma y continente
continent_platform_counts = countries_expanded.groupby(['continent', 'platform']).size().reset_index(name='count')

# 3. Filtrar los continentes con más contenido
top_continents = continent_platform_counts.groupby('continent')['count'].sum().nlargest(10).index
filtered_data = continent_platform_counts[continent_platform_counts['continent'].isin(top_continents)]

# 4. Gráfico de barras apiladas
plt.figure(figsize=(14, 8))
sns.barplot(data=filtered_data, x='continent', y='count', hue='platform')
plt.title('Distribución de Contenido por Plataforma en los Continentes con Más Títulos')
plt.xlabel('Continente')
plt.ylabel('Cantidad de Títulos')
plt.xticks(rotation=45)
plt.legend(title='Plataforma')
plt.show()
```
