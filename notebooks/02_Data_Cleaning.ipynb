```python
# Diccionario para mapear países a continentes con los nombres especificados
country_to_continent = {
    # América
    'US': 'América', 'CA': 'América', 'MX': 'América', 'BR': 'América', 'AR': 'América', 'CO': 'América',
    'CL': 'América', 'PE': 'América', 'VE': 'América', 'UY': 'América', 'PY': 'América', 'BO': 'América',
    'EC': 'América', 'CU': 'América', 'DO': 'América', 'SV': 'América', 'GT': 'América', 'HN': 'América',
    'NI': 'América', 'PA': 'América', 'CR': 'América', 'BZ': 'América',
    # Europa
    'DE': 'Europa', 'FR': 'Europa', 'GB': 'Europa', 'IT': 'Europa', 'ES': 'Europa', 'RU': 'Europa',
    'NL': 'Europa', 'BE': 'Europa', 'LU': 'Europa', 'IE': 'Europa', 'PT': 'Europa', 'CZ': 'Europa',
    'PL': 'Europa', 'RO': 'Europa', 'HU': 'Europa', 'BG': 'Europa', 'DK': 'Europa', 'SE': 'Europa',
    'NO': 'Europa', 'FI': 'Europa', 'AT': 'Europa', 'CH': 'Europa', 'GR': 'Europa', 'EE': 'Europa',
    'LT': 'Europa', 'LV': 'Europa',
    # Asia
    'CN': 'Asia', 'JP': 'Asia', 'IN': 'Asia', 'KR': 'Asia', 'TR': 'Asia', 'IQ': 'Asia', 'IR': 'Asia',
    'PH': 'Asia', 'VN': 'Asia', 'TH': 'Asia', 'MY': 'Asia', 'SG': 'Asia', 'ID': 'Asia', 'PK': 'Asia',
    'AF': 'Asia', 'BD': 'Asia', 'KH': 'Asia',
    # Oceanía
    'AU': 'Oceanía', 'NZ': 'Oceanía', 'FJ': 'Oceanía', 'PG': 'Oceanía',
    # África
    'ZA': 'África', 'EG': 'África', 'NG': 'África', 'DZ': 'África', 'MA': 'África', 'ET': 'África',
    'GH': 'África', 'KE': 'África', 'CI': 'África', 'TZ': 'África', 'UG': 'África', 'ZM': 'África',
    'ZW': 'África', 'ML': 'África', 'SN': 'África', 'SD': 'África', 'TN': 'África',
}

# Función para mapear países a continentes
def map_to_continents(countries_column):
    # Dividir los países en listas y mapear cada uno al continente
    continents = (
        countries_column.str.split(',')
        .explode()
        .str.strip()
        .map(country_to_continent)
    )
    # Agrupar por fila original y unir continentes únicos
    grouped_continents = continents.groupby(level=0).apply(lambda x: ', '.join(sorted(set(x.dropna()))))
    return grouped_continents

# Suponiendo que el DataFrame se llama toda_data_0 y tiene una columna "Países disponibles"
toda_data_0['Continentes'] = map_to_continents(toda_data_0['Países disponibles'])

column_order = ['Plataforma', 'Título', 'Tipo', 'Géneros', 'Año de lanzamiento', 'ID de IMDb', 'Calificación promedio de IMDb', 
                'Número de votos en IMDb', 'Países disponibles', 'Continentes']
toda_data_0 = toda_data_0[column_order]
```

```python
# Paso 1: Eliminar las filas donde falta el título
Toda_Data_Limpia = toda_data_0.dropna(subset=["Título"])

# Paso 2: Eliminar títulos repetidos
Toda_Data_Limpia = Toda_Data_Limpia.drop_duplicates(subset=["Título"])

# Paso 3: Eliminar las filas donde falta el género
Toda_Data_Limpia = Toda_Data_Limpia.dropna(subset=["Géneros"])

# Paso 4: Eliminar la columna "ID de IMDb"
Toda_Data_Limpia = Toda_Data_Limpia.drop(columns=["ID de IMDb"], errors="ignore")

# Paso 5: Eliminar los datos de "Año de lanzamiento" posteriores a 2024 y anteriores a 2015
Toda_Data_Limpia = Toda_Data_Limpia[
    (Toda_Data_Limpia["Año de lanzamiento"] <= 2024) &
    (Toda_Data_Limpia["Año de lanzamiento"] >= 2015)
]

# Paso 6: Rellenar valores nulos en "Calificación promedio de IMDb" y "Número de votos en IMDb"
Toda_Data_Limpia["Calificación promedio de IMDb"] = Toda_Data_Limpia["Calificación promedio de IMDb"].fillna(
    Toda_Data_Limpia["Calificación promedio de IMDb"].median()
)
Toda_Data_Limpia["Número de votos en IMDb"] = Toda_Data_Limpia["Número de votos en IMDb"].fillna(
    Toda_Data_Limpia["Número de votos en IMDb"].median()
)

# Guardar el archivo limpiado
output_path = '../data/processed/Toda_Data_Limpia.csv'
Toda_Data_Limpia.to_csv(output_path, index=False)

Toda_Data_Limpia
```
